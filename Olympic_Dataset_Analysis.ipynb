**INTRODUCTION:**

The Olympic Games are celebrated as a Grand Festival of Sports events every fourth year. This is a Golden Chance for every athletes participating in this event. The modern Olympic Games or Olympics are leading international sporting events featuring both summer and winter sports competitions in which thousands of athletes from around the world participate in a variety of competitions. The Olympic Games are considered the world's foremost sports competition with more than 200 nations participating. The Olympic Games are normally held every four years but the Summer Olympics are held over four-year period of time and the Winter Olympics is held over two-year period of the time.

Every alternate years differnet country hosts the Olympic Games. The last Summer Olympics was held at the city of Tokyo in the year 2021. 
Australia, France, Germany, Greece and Japan all hosted the Summer Olympic Games twice (with France and Australia planned to host in 2024 and 2032, respectively, taking both countries to three each).

* The data used in this Analysis is taken from Kaggle. This is 120 years of dataset from first Olympic games to 2016 Rio Olympics from the year 1896-2016.
This notebook serves to analyze and visualize the every angle of the data to help us understand the dataset and gain some information from it.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
athletes = pd.read_csv('C:/Users/91706/Olympic dataset/athlete_events.csv')
region = pd.read_csv('C:/Users/91706/Olympic dataset/noc_regions.csv')
```

```python
athletes.head()
```

```python
region.head()
```

```python
#Merging the both csv files
athletes_df = athletes.merge(region, how = 'left', on = 'NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
athletes_df.rename(columns={'region':'Region', 'notes':'Notes'}, inplace = True)
athletes_df.head()
```

```python
athletes.describe()
```

```python
nan_values = athletes_df.isna()
nan_columns = nan_values.any()
nan_columns
```

# 1) Top 10 countries with highest participation:

```python
# Top countries participating

top_10_countries= athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Top 10 countries plotting

plt.figure(figsize=(10,5))
#plt.xticks(rotation=20)
plt.title('Overall participation by countries')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette = 'Set2')
```

# 2) Age distribution of athletes:

As we can see in the histogram which age group participation is more in the Olympics. The highest participation age group are between 20-30.

```python
plt.figure(figsize=(15,6))
plt.title("Age distribution of the athletes")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color='magenta', edgecolor='white');
```

# 3) Total number of male and female participation:

# 4) Increase in participation of female athetes every year:

```python
# Total number of female athletes in each olympics

female_participants= athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex', 'Year']]
female_participants= female_participants.groupby('Year').count().reset_index()
female_participants.tail()                                 
                                 
                                  
```

* Gradual increase of womens participation in the Olympics. 

```python
womenOlympics = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data = womenOlympics, palette="Spectral")
plt.title('Women Participation in the Olympics')
```

As you can see the highest number of women participating is in the year 2016 in '**Rio Olympics**'.

* Line graph:

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over the years')
```

Since the beginnig of the Olympics there is a continuos increase in the women's participition. Although there is a little decrease around the year 1960 and 1980. But you can see from the year 1980 there's been a continuos growth in the women's participation till the year 2016.

* Gold winnig Athletes:

```python
# Gold medals athletes

goldMedals = athletes_df[(athletes_df.Medal=='Gold')]
goldMedals.head()
```

```python
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

# 5) Gold Medals won by the athletes who are more than 55 years old:

```python
# Gold won beyond the age 55

goldMedals['ID'][goldMedals['Age']>55].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age']>55]
sporting_event.head()
```

```python
# Plot for sporting_event

sns.set(style="darkgrid")
plt.figure(figsize=(12,6))
plt.tight_layout()
sns.countplot(sporting_event, palette="Set1")
plt.title('Gold Medals for Athletes over 55 years')
```

As we can see the highest number of Gold medals won by the age group of 55 plus is from Shooting, 6 Gold medals.
Total 8 sports event have gold medalist athletes over the age of 55 years in the Olympics.

* **Alpinism, Art competiton, Roque & Croquet all these event was abandoned before mid 1950's. So, it is no longer played in the Olympics.**

# 5) Total number of Gold medal won by each country:

```python
# Number of Gold Medals from each country

goldMedals.Region.value_counts().reset_index(name='Medal')
```

* We can see clearly that United Satets is no. 1 in this race and they have the highest number of Gold medals counts and they are very ahead from its competitors with a very good lead.

  There are many countries with one Gold medal.

```python
# Plotting Gold medals per country

totalGoldmedals = goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x="index", y="Medal", data = totalGoldmedals, height=5, kind="bar", palette="magma")
g.despine(left=True)
g.set_xlabels("Top 6 countries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medals per Country')
```

# 6) **Rio Summer Olympics 2016:**

```python
# Rio Oylmpics
max_year = athletes_df.Year.max()
print("Rio Olympics - 2016")
team_names = athletes_df[(athletes_df.Year== max_year) & (athletes_df.Medal=='Gold')].Team

team_names.value_counts().head(10)
```
