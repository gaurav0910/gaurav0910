```python
import json
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
scrape_url = "https://understat.com/league/La_liga/2020"
```

```python
urlopen(scrape_url)
```

```python
page_connect = urlopen(scrape_url)
```

```python
page_connect
```

```python
page_html = BeautifulSoup(page_connect, "html.parser")
```

```python
page_html
```

```python
page_html.findAll(name="script")[3]
```

```python
page_html.findAll(name="script")[3].text
```

```python
json_string = page_html.findAll(name="script")[3].text
```

```python
json_string
```

```python
start_ind = json_string.index("\\")
end_ind = json_string.index("')")
```

```python
json_string[start_ind:end_ind]
```

```python
json_data = json_string[start_ind:end_ind]
```

```python
json_data.encode("utf8").decode("unicode_escape")
```

```python
json_data = json_data.encode("utf8").decode("unicode_escape")
```

```python
json.loads(json_data)
```

```python
final_json_df = pd.json_normalize(json.loads(json_data))
final_json_df
```
